```python
import pandas as pd
```

```python
# Load the CSV file
file_path = 'vehicle_cleaned.csv'
data = pd.read_csv(file_path)
```

```python
# Extract unique zip codes
unique_zip_codes = data['zip_code'].unique()

# Convert to DataFrame for better display
unique_zip_codes_df = pd.DataFrame(unique_zip_codes, columns=['Unique Zip Codes'])

# Display the unique zip codes
unique_zip_codes_df.head()
```

```python
# Aggregate the 'vehicles' column by 'zip_code'
aggregated_data = data.groupby('zip_code', as_index=False).agg({
    'vehicles': 'sum',
    'lat': 'mean',
    'lon': 'mean' 
})
```

```python
# Save the aggregated data to a new CSV file
aggregated_data.to_csv('static/data/aggregated_vehicle_data.csv', index=False)

# Display the aggregated data
aggregated_data
```
